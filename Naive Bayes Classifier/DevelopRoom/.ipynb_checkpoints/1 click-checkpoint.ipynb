```python
import numpy as np
import matplotlib.pyplot as plt
from random import randint
from sklearn.utils import shuffle
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import time
from math import fabs
from sklearn.metrics import confusion_matrix

while True:
    noise = np.random.randn(997, 2)
    
    cloud1 = noise
    cloud2 = noise + np.array([randint(-2, 2),randint(-2, 2)])
    cloud3 = noise + np.array([randint(-2, 2),randint(-2, 2)])
    cloud4 = noise + np.array([randint(-2, 2),randint(-2, 2)])
    cloud5 = noise + np.array([randint(-2, 2), randint(-2, 2)])

    X = np.vstack([cloud1, cloud2, cloud3, cloud4, cloud5])

    annot1 = [0] * cloud1.shape[0]
    annot2 = [1] * cloud2.shape[0]
    annot3 = [2] * cloud3.shape[0]
    annot4 = [3] * cloud4.shape[0]
    annot5 = [4] * cloud5.shape[0]
    y = annot1 +annot2 + annot3 + annot4 + annot5

    data, labels = shuffle(X, y)

    test_size = int(len(data) * 0.5)
    Xtrain = data[:test_size]
    Ytrain = labels[:test_size]
    Xtest = data[test_size:]
    Ytest = labels[test_size:]

    clf = LogisticRegression()
    clf.fit(Xtrain, Ytrain)

    predictions = clf.predict(Xtest)

    if accuracy_score(Ytest, predictions)>=0.704:
        if accuracy_score(Ytest, predictions)<=0.716:
            break
accuracy_score(Ytest, predictions)

accuracy = [0,0,0,0,0]
times = [0,0,0,0,0]
var = [0,0,0,0,0]
matrix = [0,0,0,0,0]

dict = [[],[],[],[],[]]
i = 0 
for element in Xtrain:
    classs = Ytrain[i] # переменная classs нужна для того, чтобы каждый элемент знал в какой из подмассивов
                       # он отправится
    dict[classs].append(element) # здесь как раз сортируются координаты по соответствующим подмассивам
    i+=1
    
def train (mass):
    allx = []
    ally = []
    for k in range(0, len(mass)):
        allx.append(mass[k][0])
        ally.append(mass[k][1])
    averx = sum(allx)/len(mass)
    avery = sum(ally)/len(mass)
    return (averx, avery)

avx1, avy1 = train(dict[0])
avx2, avy2 = train(dict[1])
avx3, avy3 = train(dict[2])
avx4, avy4 = train(dict[3])
avx5, avy5 = train(dict[4])

def predict (test):
    predict = []
    list = []
    start = time.time()
    for co in test:
        inner_start = time.time()
        diffX1 = fabs(co[0]-avx1)
        diffX2 = fabs(co[0]-avx2)
        diffX3 = fabs(co[0]-avx3)
        diffX4 = fabs(co[0]-avx4)
        diffX5 = fabs(co[0]-avx5)
        diffY1 = fabs(co[1]-avy1)
        diffY2 = fabs(co[1]-avy2)
        diffY3 = fabs(co[1]-avy3)
        diffY4 = fabs(co[1]-avy4)
        diffY5 = fabs(co[1]-avy5)
        
        pred1 = diffX1 * diffY1
        pred2 = diffX2 * diffY2
        pred3 = diffX3 * diffY3
        pred4 = diffX4 * diffY4
        pred5 = diffX5 * diffY5
        
        min = pred1
        clf = 0
        if pred2<min:
            min = pred2
            clf = 1
        if pred3<min:
            min= pred3
            clf = 2
        if pred4<min:
            min= pred4
            clf = 3
        if pred5<min:
            min= pred5
            clf = 4
        predict.append(clf)
        inner_times = time.time() - inner_start
        list.append(inner_times)
    TIME = time.time() - start
    VAR = np.var(list)
    return (predict, TIME, VAR)

result = predict(Xtest)

predict = result[0]
clfTIME = result[1]/len(Ytest)
clfVAR = result[2]
clfCONFMAT = confusion_matrix(Ytest,predict)
clfACC = accuracy_score(Ytest, predict)

accuracy[4] = clfACC
times[4] = clfTIME
var[4] = clfVAR
matrix[4] = clfCONFMAT

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
Xtrain = scaler.fit_transform(Xtrain)
Xtest = scaler.fit_transform(Xtest)

from sklearn.naive_bayes import GaussianNB

def gaus (Xtrain, Ytrain, Xtest, Ytest):
    list =[]
    clf = GaussianNB()
    clf.fit(Xtrain, Ytrain)
    
    while True:
        start = time.time()
        predict = clf.predict(Xtest)
        TIME = time.time() - start
        TIME = TIME/len(Ytest)
        if TIME > 0:
            break
    
    for i in Xtest:
        tmp = [[],[]]
        tmp[0].append(i[0])
        tmp[1].append(i[1])
        start = time.time()
        pred = clf.predict(tmp)
        TIMES = time.time() - start
        list.append(TIMES)
        VAR = np.var(list)
        
    ACC = accuracy_score(Ytest, predict)
    CONFMAT = confusion_matrix(Ytest,predict)
    return (ACC, CONFMAT, TIME, VAR)

gausACC, gausCONFMAT, gausTIME, gausVAR = gaus (Xtrain, Ytrain, Xtest, Ytest)

accuracy[0] = gausACC
times[0] = gausTIME
var[0] = gausVAR
matrix[0] = gausCONFMAT

from sklearn.naive_bayes import MultinomialNB

def multi (Xtrain, Ytrain, Xtest, Ytest):
    list =[]
    clf = MultinomialNB()
    clf.fit(Xtrain, Ytrain)
    
    while True:
        start = time.time()
        predict = clf.predict(Xtest)
        TIME = time.time() - start
        TIME = TIME/len(Ytest)
        if TIME > 0:
            break
        
        for i in Xtest:
            tmp = [[],[]]
            tmp[0].append(i[0])
            tmp[1].append(i[1])
    
            start = time.time()
            pred = clf.predict(Xtest)
            TIMES = time.time() - start
            list.append(TIMES)
            VAR = np.var(list)
        
    ACC = accuracy_score(Ytest, predict)
    CONFMAT = confusion_matrix(Ytest,predict)
    return (ACC, CONFMAT, TIME, VAR)

multiACC, multiCONFMAT, multiTIME, multiVAR = multi(Xtrain, Ytrain, Xtest, Ytest)

accuracy[1] = multiACC
times[1] = multiTIME
var[1] = multiVAR
matrix[1] = multiCONFMAT

from sklearn.naive_bayes import ComplementNB

def compl (Xtrain, Ytrain, Xtest, Ytest):
    list =[]
    clf = ComplementNB()
    clf.fit(Xtrain, Ytrain)
    
    while True:
        start = time.time()
        predict = clf.predict(Xtest)
        TIME = time.time() - start
        TIME = TIME/len(Ytest)
        if TIME > 0:
            break
        
        for i in Xtest:
            tmp = [[],[]]
            tmp[0].append(i[0])
            tmp[1].append(i[1])
            start = time.time()
            pred = clf.predict(Xtest)
            TIMES = time.time() - start
            list.append(TIMES)
            VAR = np.var(list)
        
    ACC = accuracy_score(Ytest, predict)
    CONFMAT = confusion_matrix(Ytest,predict)
    return (ACC, CONFMAT, TIME, VAR)

complACC, complCONFMAT, complTIME, complVAR = compl(Xtrain, Ytrain, Xtest, Ytest)

accuracy[2] = complACC
times[2] = complTIME
var[2] = complVAR
matrix[2] = complCONFMAT

from sklearn.naive_bayes import BernoulliNB

def bern (Xtrain, Ytrain, Xtest, Ytest):
    list =[]
    clf = BernoulliNB()
    clf.fit(Xtrain, Ytrain)
    
    while True:
        start = time.time()
        predict = clf.predict(Xtest)
        TIME = time.time() - start
        TIME = TIME/len(Ytest)
        if TIME > 0:
            break
        
        for i in Xtest:
            tmp = [[],[]]
            tmp[0].append(i[0])
            tmp[1].append(i[1])
            start = time.time()
            pred = clf.predict(Xtest)
            TIMES = time.time() - start
            list.append(TIMES)
            VAR = np.var(list)
        
    ACC = accuracy_score(Ytest, predict)
    CONFMAT = confusion_matrix(Ytest,predict)
    return (ACC, CONFMAT, TIME, VAR)

bernACC, bernCONFMAT, bernTIME, bernVAR = bern(Xtrain, Ytrain, Xtest, Ytest)

accuracy[3] = bernACC
times[3] = bernTIME
var[3] = bernVAR
matrix[3] = bernCONFMAT

import seaborn as sns
clfNAME = ['Gaussian', 'Multinomial', 'Complement', 'Bernoulli', 'MyClf']

fig, axes = plt.subplots(1, 3,sharex=True, figsize=(16,8))
fig.suptitle('Comparing naive classificators')
axes[0].set_title('Accuracy')
axes[1].set_title('Time')
axes[2].set_title('Var')

sns.barplot(ax=axes[0], x=clfNAME, y=accuracy)

sns.barplot(ax=axes[1], x=clfNAME, y=times)

sns.barplot(ax=axes[2], x=clfNAME, y=var)
```
